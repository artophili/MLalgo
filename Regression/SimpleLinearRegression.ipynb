Simple linear regression tries to fit a straight line
The condition is do draw the line is minimize the error

error = summation of (delta)^2
Here delta : Distance between the point and line drawn
then we sum up all the deltas to find the overall error and select that line which have minimum error

It follows the line's slope point equation.
y = m*x + b
where m : slope/gradient
b : y intercept

Goal : Predict the price of house based on the area
Equation become:
price = m * area + b
Here price is a dependent variable and
area is independent variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area in Sq.Ft')
plt.ylabel('Price in $')
plt.scatter(df['area'],df['price'], color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area']],df.price)
```

```python
x_test = np.array([[3300]])
```

```python
reg.predict(x_test)
```

```python
#Predict function requires 2D array 
#The predict function requires the input shape to be 
#(n_samples,n_features)
#Even if there is only one sample and one feature
#If we want to predict the value for only one sample point then 
#we have to convert that into 2d numpy array using np.array([[]]) method
```

```python
m = reg.coef_
```

```python
b = reg.intercept_
```

```python
area = int(input("Enter an area to check the house price:"))

price = m * area + b
print(price)
```

```python
d = pd.read_csv('areas.csv')
d.head()
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
```

```python
d.head()
```

```python
d.to_csv('predictions.csv')
```
