```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
#Creating a blob of 200 data points

dataset = make_blobs(n_samples=200,
                     n_features=2,
                     centers=4,
                     cluster_std=1.6,
                     random_state=41)
```

```python
#The above function creates a array which have 2 elements
points = dataset[0]
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
#Creating a dendrogram
dendrograms = sch.dendrogram(sch.linkage(points, method='ward'))
```

```python
plt.scatter(dataset[0][:,0], dataset[0][:,1])
```

```python
#Perform the actual clustering

hc = AgglomerativeClustering(n_clusters=4, linkage='ward')
```

```python
y_hc = hc.fit_predict(points)
```

```python
plt.scatter(points[y_hc==0,0], points[y_hc==0,1], s=100, c='cyan')
plt.scatter(points[y_hc==1,0], points[y_hc==1,1], s=100, c='yellow')
plt.scatter(points[y_hc==2,0], points[y_hc==2,1], s=100, c='green')
plt.scatter(points[y_hc==3,0], points[y_hc==3,1], s=100, c='red')
```
